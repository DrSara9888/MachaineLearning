```python
import pandas as pd 
```

```python
df= pd.read_csv("/Users/khadeejah/Desktop/Decision_Tree/titanic.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df.head(10)
```

```python
inputs = df.drop(['Survived','Ticket','Cabin','Embarked','SibSp','Parch'],axis='columns')
```

```python
target = df['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex=LabelEncoder()
le_name=LabelEncoder()
```

```python
inputs['sex_n']=le_sex.fit_transform(inputs['Sex'])
inputs['name_n']=le_sex.fit_transform(inputs['Name'])
```

```python
inputs
```

```python
from sklearn import tree
```

```python
inputs_in=inputs.drop(['Sex','Name','name_n','PassengerId'],axis='columns')

inputs_in.head()
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_in, target)
```

```python
model.score(inputs_in,target)
```
